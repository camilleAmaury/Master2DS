### Mushroom dataset

```python
import pandas as pd
data = pd.read_excel('Mushroom.xlsx')
data.head()
```

```python
rules, supp_dict = ar.create_association_rules(data.drop('Id',axis=1),max_cols=2,binning_method=five_quantile_based_bins)
```

```python
rules.head()
```

```python
rules[(rules['confidence'] > 0.7) & (rules['lift'] > 1) ].head()
```

As another example, we can from a smelly mushroom expect a large foot mushroom aswell with a 92% probability

Both the example you saw are made only with 1 antecedent and consequent. Let's now allow the association rules to contain more antecedents and consequents.

```python
rules, supp_dict = ar.create_association_rules(data.drop('Id',axis=1),max_cols=4,binning_method=five_quantile_based_bins)
rules.head()
```

```python
rules[(rules['confidence'] > 0.7) & (rules['lift'] > 1.2)].head()
```

We now have a lot more association rules to choose from,  but as you can see, as the dataset is small a alot of these probabilities are equal to one because there is only one example with the combination of antecedents and consequents.

```python
from sklearn.datasets import load_iris
data = load_iris()
df_iris = pd.DataFrame(data.data, columns=data.feature_names)
df_iris.head()
rules, supp_dict = ar.create_association_rules(df_iris,binning_method=five_quantile_based_bins)
rules.head()
```

The same exercice on the iris dataset becomes a lot more relevant for two reasons:
* We are working with ranges of number, which allows us to cover relations with more accuracy.
* The dataset doesn't contain a dozen of lines but 10, which allows more reliable values.

Also, this kind of dataset is able to give rules not only between the features and a class but between features themselves.

```python
rules[(rules['confidence'] > 0.8) & (rules['lift'] > 1.2) & (rules['ant_count'] > 4)].head()
```

```python
target = rules[(rules['confidence'] > 0.8) & (rules['lift'] > 1.2) & (rules['ant_count'] > 4)].iloc[0]
print(target['antecedent'], target['consequent'], target['confidence'])
```

In this example we can see that a given width for both petals and sepals of an iris, within a range, can give the approximate length of their petals and sepals with a 83% chance, which is a very correct score considering the dataset size.
