### Probability of Occurences

```python
PA = pd.DataFrame([[25/50, 25/50]], columns=[0, 1], index=["P(A)"])
PA.head()
```

```python
PD = pd.DataFrame([[26/50, 24/50]], columns=[0, 1], index=["P(D)"])
PD.head()
```
