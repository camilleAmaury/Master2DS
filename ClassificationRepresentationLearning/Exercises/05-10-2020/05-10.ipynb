## Ceasar Cypher

![Ex1](1.PNG)

```python
class CeasarCypher(object):
    def __init__(self):
        self._code = {
            'a':'n','b':'o','c':'p','d':'q','e':'r','f':'s','g':'t','h':'u',
            'i':'v','j':'w','k':'x','l':'y','m':'z','n':'a','o':'b','p':'c','q':'d',
            'r':'e','s':'f','t':'g','u':'h','v':'i','w':'j','x':'k','y':'l','z':'m',
            'A':'N','B':'O','C':'P','D':'Q','E':'R','F':'S','G':'T','H':'U','I':'V',
            'J':'W','K':'X','L':'Y','M':'Z','N':'A','O':'B','P':'C','Q':'D','R':'E',
            'S':'F','T':'G','U':'H','V':'I','W':'J','X':'K','Y':'L','Z':'M'
        }
        self._invcode = {v: k for k, v in self._code.items()}
    
    def encrypt(self, message):
        return "".join([(self._code[i] if i.isalpha() else i) for i in message])
    
    def decrypt(self, encrypted):
        return "".join([(self._invcode[i] if i.isalpha() else i) for i in encrypted])
```

```python
cc = CeasarCypher()
```

```python
to_encrypt = "but nonetheless necessary"
print("Message To encrypt ({}) ==> ({})".format(to_encrypt, cc.encrypt(to_encrypt)))
print("Decrypted == initial message ? {}".format(cc.decrypt(cc.encrypt(to_encrypt)) == to_encrypt))
```

```python
to_decrypt = "BZT! guvf vf fb obevat"
print("Message To decrypt ({}) ==> ({})".format(to_decrypt, cc.decrypt(to_decrypt)))
```

## Numpy Matrixes

![Ex2](2.PNG)

```python
import numpy as np
```

```python
A = np.matrix([
    [2,1,1,0],
    [1,2,0,1],
    [1,2,3,2],
    [0,2,1,2]
], dtype=float)
B = np.matrix([
    [9], 
    [8], 
    [7], 
    [6]
], dtype=float)
```

### Algebra

* A * X = B
* <=> A^-1 * A * X = A^-1 * B 
* <=> I * X = A^-1 * B 
* <=> X = A^-1 * B 

```python
X = np.dot(np.linalg.inv(A), B)
X
```

```python
print("AX == B ? {}".format(np.allclose(np.dot(A, X),B)))
```

```python
eig = np.linalg.eigh(A)
print("> Eigen Values :\n   > {}\n> Eigen Vectors :\n{}".format(eig[0], eig[1]))
```

```python
nb_points = 200
def f(x,y): return (1-2*x+x**3+2*y**3)*np.exp(-x**2-y**2)

n = 20
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
plt.show()
```
